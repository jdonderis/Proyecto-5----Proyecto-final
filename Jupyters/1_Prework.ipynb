A continuación, vamos a juntar los 5 datasets en uno solo a través de las columnas player_id y person_id ya que parece que contienen los id de todos los jugadores. Para ello vamos a seguir los siguientes pasos:

***Paso 1 --> Normalizamos el nombre de las columnas player_id y person_id***

Para normalizarla vamos a hacer uso de una función previamente definida con la misión de renombrar columnas: 

```python
common_info = renombrar_columnas (common_info, 'person_id', 'id')
common_info.columns
```

```python
draft_history = renombrar_columnas(draft_history, 'person_id', 'id')
draft_history.columns
```

```python
players.columns
```

```python
players_general = renombrar_columnas(players_general, 'player_id', 'id')
players_general.columns
```

***Paso 2 --> Unimos las tablas***

Uniremos la tabla de players_general con el resto de tablas para conservar el máximo de información posible 

```python
#Unimos las dos primeras tablas
player_def = pd.merge(players_general, players, on= 'id', how= 'left')
player_def.columns
```

```python
#Unimos ahora la información histórica del draft
players_def2 = pd.merge(player_def, draft_history, on= 'id', how= 'left')
player_def.columns
```

Para unificar la última base de datos tenemos que realizar una limpieza ya que hay varias columnas repetidas como podemos observar. 

```python
players_def2 = players_def2.drop(columns = ['first_name_x', 'last_name_x', 'first_name_y','last_name_y'])
players_def2.columns
```

```python
#Por último, unificamos los datos de common player info
players_definitivo = pd.merge(players_def2, common_info, on = 'id', how = 'left')
players_definitivo.columns
```

```python
#A continuación observamos el tamaño y los registros en la base de datos
players_definitivo.shape
```
